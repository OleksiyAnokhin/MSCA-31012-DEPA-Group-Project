# Google Mobility Data

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading and Transforming data

```python
google_data=pd.read_csv('Data/Global_Mobility_Report.csv',parse_dates=True,index_col='date')
df=pd.DataFrame(google_data)
df_us_il_cookcounty=df[(df['country_region']=='United States') & (df['sub_region_1']=='Illinois')& (df['sub_region_2']=='Cook County')]
df_us_il_cookcounty.head()
```

```python
fig,ax=plt.subplots()
plt.style.use('seaborn')
ax.plot(df_us_il_cookcounty.index,df_us_il_cookcounty['retail_and_recreation_percent_change_from_baseline'],color='mediumaquamarine')
ax.set_xlabel('Dates')
ax.set_ylabel('Percentage Change')
fig.suptitle('Retail and recreation', fontsize=14, fontweight='bold')
ax.tick_params(labelrotation=45)
plt.show()
fig.savefig('Google_retail.png')
```

**The Retail and Recreation has dropped drastically from the baseline since Shelter in place order from 20/03/2020, and hence this isnt an arena we need to focus on in terms of reopening because the drop states it not being a necessity**

```python
fig,ax=plt.subplots()
plt.style.use('seaborn')
ax.plot(df_us_il_cookcounty.index,df_us_il_cookcounty['grocery_and_pharmacy_percent_change_from_baseline'],color='mediumaquamarine')
ax.set_xlabel('Dates')
ax.set_ylabel('Percentage Change')
fig.suptitle('Grocery and pharmacy', fontsize=14, fontweight='bold')
ax.tick_params(labelrotation=45)
plt.show()
fig.savefig('Google_grocery.png')
```

**We have a spike in Groceries just before Shelter in Place was announced. We expect this because many people stocked food for months to come. Since then the activity is closest to reach to baseline so we would like to consider it our analysis**

```python
fig,ax=plt.subplots()
plt.style.use('seaborn')
ax.plot(df_us_il_cookcounty.index,df_us_il_cookcounty['parks_percent_change_from_baseline'],color='mediumaquamarine')
ax.set_xlabel('Dates')
ax.set_ylabel('Percentage Change')
fig.suptitle('Parks', fontsize=14, fontweight='bold')
ax.tick_params(labelrotation=45)
fig.savefig('Google_parks.png')
```

**We can see from the graphs that since Shelter in Place was announced mobility in national parks, public beaches, marinas, dog parks, plazas, and public gardens has increased considerably which make them a probable next hotspots for COVID. Hence, we would like to consider them in our analysis**

```python
fig,ax=plt.subplots()
plt.style.use('seaborn')
ax.plot(df_us_il_cookcounty.index,df_us_il_cookcounty['transit_stations_percent_change_from_baseline'],color='mediumaquamarine')
ax.set_xlabel('Dates')
ax.set_ylabel('Percentage Change')
fig.suptitle('Transit stations', fontsize=14, fontweight='bold')
ax.tick_params(labelrotation=45)
fig.savefig('Google_transit.png')
```

**In terms of reopening the city, we expect public transit in Chicago to play a major role because startegically opening the routes and stations to the public is the only way to avoid a spike in number of COVID cases** 

```python
fig,ax=plt.subplots()
plt.style.use('seaborn')
ax.plot(df_us_il_cookcounty.index,df_us_il_cookcounty['workplaces_percent_change_from_baseline'],color='mediumaquamarine')
ax.set_xlabel('Dates')
ax.set_ylabel('Percentage Change')
fig.suptitle('Workplace', fontsize=14, fontweight='bold')
ax.tick_params(labelrotation=45)
fig.savefig('Google_workplace.png')
```

**Major companies are shifting to the practice of work from home. We assume the practice to go on because there have been reports where companies are offering this practice even after the reopening, and hence we aren't considering this in our analsysis**

```python
fig,ax=plt.subplots()
plt.style.use('seaborn')
ax.plot(df_us_il_cookcounty.index,df_us_il_cookcounty['residential_percent_change_from_baseline'],color='mediumaquamarine')
ax.set_xlabel('Dates')
ax.set_ylabel('Percentage Change')
fig.suptitle('Resedential', fontsize=14, fontweight='bold')
ax.tick_params(labelrotation=45)
plt.show()
fig.savefig('Google_resedential.png')
```

**Having Mobility in resedential area isn't much of a concern because the data doesn't state that whether the mobility is contained within their homes or outside. We believe that mobility in resedential areas isnt that bad because unlike groceries, or workplaces, people are 6ft apart and maintain social distancing**
